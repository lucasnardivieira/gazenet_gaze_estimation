## Modulo 2 -- Deteccao de Pessoas e Faces

**Problema v3.0:** posicao dos olhos = 20% do topo do bbox da pessoa (heuristica grosseira).

**Solucao v4.0:** MediaPipe Face Detection localiza o rosto real;
MediaPipe Face Mesh extrai landmarks precisos dos olhos.
Inclui tambem a abordagem classica de estimacao de gaze por geometria.

### 2.1 MediaPipe: Deteccao de Rostos e Landmarks

```python
import mediapipe as mp

mp_face_detection = mp.solutions.face_detection
mp_face_mesh      = mp.solutions.face_mesh

LEFT_EYE_CENTER  = [33,  133]
RIGHT_EYE_CENTER = [362, 263]

def get_face_and_eyes_mediapipe(img_pil):
    img_rgb = np.array(img_pil)
    W, H = img_pil.size
    detections = []

    with mp_face_detection.FaceDetection(
            model_selection=1, min_detection_confidence=0.4) as fd:
        result_det = fd.process(img_rgb)
    if not result_det.detections:
        return detections

    with mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=10,
            refine_landmarks=True, min_detection_confidence=0.4) as fm:
        result_mesh = fm.process(img_rgb)

    lm_list = result_mesh.multi_face_landmarks if result_mesh.multi_face_landmarks else []

    for i, det in enumerate(result_det.detections):
        bb = det.location_data.relative_bounding_box
        fx = max(0.0, bb.xmin); fy = max(0.0, bb.ymin)
        fw = min(1.0-fx, bb.width); fh = min(1.0-fy, bb.height)

        if i < len(lm_list):
            lm  = lm_list[i].landmark
            lx  = float(np.mean([lm[j].x for j in LEFT_EYE_CENTER]))
            ly  = float(np.mean([lm[j].y for j in LEFT_EYE_CENTER]))
            rx  = float(np.mean([lm[j].x for j in RIGHT_EYE_CENTER]))
            ry  = float(np.mean([lm[j].y for j in RIGHT_EYE_CENTER]))
            eye_x, eye_y = (lx+rx)/2, (ly+ry)/2
            success = True
        else:
            eye_x, eye_y = fx+fw/2, fy+fh*0.35
            lx, ly, rx, ry = eye_x, eye_y, eye_x, eye_y
            success = False

        detections.append({
            "face_id":i+1,
            "face_bbox_norm":[fx,fy,fw,fh],
            "face_bbox_pixel":[int(fx*W),int(fy*H),int((fx+fw)*W),int((fy+fh)*H)],
            "eye_x":eye_x,"eye_y":eye_y,
            "left_eye":(lx,ly),"right_eye":(rx,ry),
            "success":success,
        })
    return detections

print(f"MediaPipe {mp.__version__} -- funcoes definidas!")
```

### 2.2 Deteccao de Pessoas (YOLOv8) + Faces (MediaPipe)

```python
from ultralytics import YOLO
yolo_model = YOLO("yolov8m.pt")
PERSON_CLASS_ID = 0

def detect_people_and_faces(image_path, conf_threshold=0.4):
    results_yolo = yolo_model(image_path, conf=conf_threshold, verbose=False)[0]
    img = Image.open(image_path).convert("RGB")
    W, H = img.size
    all_faces = get_face_and_eyes_mediapipe(img)
    people = []
    pid = 0
    for box in results_yolo.boxes:
        if int(box.cls[0].item()) != PERSON_CLASS_ID: continue
        pid += 1
        x1,y1,x2,y2 = box.xyxy[0].cpu().numpy()
        conf = float(box.conf[0].item())
        xn1,yn1,xn2,yn2 = x1/W,y1/H,x2/W,y2/H

        face_match = None; best_area = 0.0
        for face in all_faces:
            fx,fy,fw,fh = face["face_bbox_norm"]
            fc_x,fc_y = fx+fw/2, fy+fh/2
            if xn1<=fc_x<=xn2 and yn1<=fc_y<=yn2:
                area = fw*fh
                if area > best_area:
                    best_area=area; face_match=face

        if face_match:
            eye_x,eye_y = face_match["eye_x"],face_match["eye_y"]
            eye_src = "mediapipe"
            fx,fy,fw,fh = face_match["face_bbox_norm"]
            m=0.1
            hx1=max(0,int((fx-fw*m)*W)); hy1=max(0,int((fy-fh*m)*H))
            hx2=min(W,int((fx+fw*(1+m))*W)); hy2=min(H,int((fy+fh*(1+m))*H))
        else:
            eye_x=(xn1+xn2)/2; eye_y=yn1+(yn2-yn1)*0.20
            eye_src="fallback"
            hx1,hy1,hx2,hy2 = int(x1),int(y1),int(x2),int(y2)

        people.append({
            "person_id":pid,
            "bbox_norm":[xn1,yn1,xn2-xn1,yn2-yn1],
            "bbox_pixel":[x1,y1,x2,y2],
            "head_bbox_pixel":[hx1,hy1,hx2,hy2],
            "eye_x":eye_x,"eye_y":eye_y,
            "eye_source":eye_src,"confidence":conf,
        })

    mp_hits = sum(1 for p in people if p["eye_source"]=="mediapipe")
    if people: print(f"  {len(people)} pessoa(s) | MediaPipe: {mp_hits}/{len(people)}")
    return img, people

print("detect_people_and_faces v4.0 definida!")
```

### 2.3 Abordagem Classica: Estimacao de Gaze por Geometria

**Nucleo da comparacao academica.**
Usa apenas landmarks dos olhos para estimar a direcao do olhar -- zero treinamento.

```python
def estimate_gaze_classic(img_pil, person):
    """
    Abordagem CLASSICA via MediaPipe + geometria pura.
    1. Vetor entre os dois olhos -> orientacao horizontal da cabeca
    2. Vetor perpendicular -> direcao frontal estimada
    3. Projeta a partir do centro dos olhos

    Vantagem: zero treinamento, rapido, interpretavel.
    Limitacao: ignora contexto da cena; falha com rostos laterais/pequenos.
    """
    if person.get("eye_source") != "mediapipe":
        return None
    img_rgb = np.array(img_pil)
    with mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=10,
            refine_landmarks=True, min_detection_confidence=0.4) as fm:
        result = fm.process(img_rgb)
    if not result.multi_face_landmarks:
        return None

    lm = result.multi_face_landmarks[0].landmark
    left_eye  = np.array([np.mean([lm[i].x for i in LEFT_EYE_CENTER]),
                           np.mean([lm[i].y for i in LEFT_EYE_CENTER])])
    right_eye = np.array([np.mean([lm[i].x for i in RIGHT_EYE_CENTER]),
                           np.mean([lm[i].y for i in RIGHT_EYE_CENTER])])

    eye_center  = (left_eye + right_eye) / 2
    horiz_vec   = right_eye - left_eye
    eye_dist    = np.linalg.norm(horiz_vec) + 1e-6
    forward_vec = np.array([-horiz_vec[1], horiz_vec[0]])
    forward_vec = forward_vec / (np.linalg.norm(forward_vec) + 1e-6)
    gaze_pt     = eye_center + forward_vec * (eye_dist * 3.5)
    return float(np.clip(gaze_pt[0],0,1)), float(np.clip(gaze_pt[1],0,1))

print("estimate_gaze_classic definida!")
```

## Modulo 3 -- Dataset e DataLoaders v4.0

**Melhorias sobre v3.0:**
- `HEAD_SIZE=112` (era 64): aproveita crop facial real do MediaPipe
- `sigma=5.0` (era 3.0): targets mais suaves, condizentes com variancia humana
- Eye Encoder recebe 4 valores (dois olhos x,y)
- Multiplas anotacoes por imagem agrupadas em um unico heatmap

### 3.1 Classe do Dataset

```python
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
import torch.nn.functional as F

HEATMAP_SIZE = 64
HEAD_SIZE    = 112

class GazeFollowDataset(Dataset):
    IMG_SIZE  = 224
    HEAD_SIZE = HEAD_SIZE
    HMAP_SIZE = HEATMAP_SIZE
    SIGMA     = 5.0

    def __init__(self, df, images_root, augment=True, fraction=1.0):
        if fraction < 1.0:
            self.df = df.sample(frac=fraction, random_state=42).reset_index(drop=True)
        else:
            self.df = df.reset_index(drop=True)
        self.images_root = images_root
        self.augment     = augment
        key_cols = ["image_path","x_min","y_min","w","h","eye_x","eye_y"]
        self.grouped = list(self.df.groupby(key_cols, sort=False))

        self.scene_tf = T.Compose([
            T.Resize((self.IMG_SIZE,self.IMG_SIZE)), T.ToTensor(),
            T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])])
        self.head_tf = T.Compose([
            T.Resize((self.HEAD_SIZE,self.HEAD_SIZE)), T.ToTensor(),
            T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])])
        self.scene_aug = T.Compose([
            T.ColorJitter(0.4,0.4,0.3,0.1), T.RandomGrayscale(p=0.05),
            T.RandomApply([T.GaussianBlur(3)],p=0.2),
            T.RandomApply([T.RandomPerspective(0.1)],p=0.2)])
        self.head_aug   = T.Compose([T.ColorJitter(0.3,0.3), T.RandomGrayscale(p=0.05)])
        self.head_erase = T.RandomErasing(p=0.2, scale=(0.02,0.15))

    def _make_heatmap(self, gaze_points):
        S = self.HMAP_SIZE
        hmap = torch.zeros(S,S)
        y_g = torch.arange(S,dtype=torch.float32)
        x_g = torch.arange(S,dtype=torch.float32)
        gy, gx = torch.meshgrid(y_g, x_g, indexing='ij')
        for gaze_x, gaze_y in gaze_points:
            xi=gaze_x*S; yi=gaze_y*S
            hmap += torch.exp(-((gx-xi)**2+(gy-yi)**2)/(2*self.SIGMA**2))
        return hmap / (hmap.sum()+1e-8)

    def __len__(self): return len(self.grouped)

    def __getitem__(self, idx):
        key, group = self.grouped[idx]
        img_path,x_min,y_min,w,h,eye_x,eye_y = key
        img = Image.open(os.path.join(self.images_root, img_path.strip())).convert("RGB")
        Wi,Hi = img.size
        x1=max(0,int(x_min*Wi)); y1=max(0,int(y_min*Hi))
        x2=min(Wi,int((x_min+w)*Wi)); y2=min(Hi,int((y_min+h)*Hi))
        head_img = img.crop((x1,y1,x2,y2)) if x2>x1 and y2>y1 else img

        flip=False
        if self.augment:
            if torch.rand(1).item()<0.5:
                img=img.transpose(Image.FLIP_LEFT_RIGHT)
                head_img=head_img.transpose(Image.FLIP_LEFT_RIGHT); flip=True
            img=self.scene_aug(img); head_img=self.head_aug(head_img)

        scene_t=self.scene_tf(img); head_t=self.head_tf(head_img)
        if self.augment: head_t=self.head_erase(head_t)

        ex = 1.0-eye_x if flip else eye_x
        off = 0.03
        eye_input = torch.tensor([max(0.,ex-off),eye_y,min(1.,ex+off),eye_y],
                                  dtype=torch.float32)
        gaze_pts = [(1.0-r["gaze_x"] if flip else r["gaze_x"], r["gaze_y"])
                    for _,r in group.iterrows()]
        heatmap  = self._make_heatmap(gaze_pts)
        gaze_mean= torch.tensor([float(np.mean([g[0] for g in gaze_pts])),
                                  float(np.mean([g[1] for g in gaze_pts]))],
                                 dtype=torch.float32)
        return scene_t, head_t, eye_input, heatmap, gaze_mean

print("GazeFollowDataset v4.0 definido!")
```

## Modulo 4 -- GazeNet v4.0: Arquitetura e Treinamento

**Melhorias sobre v3.0:**
- Eye Encoder: 4 inputs (dois olhos) vs 2
- Head Encoder: 112x112 vs 64x64
- Skip connection U-Net: features 28x28 do ResNet-50 layer2 no decoder
- Loss rebalanceada: KL=2.0, L1=2.0

### 4.1 Modelo

```python
import torch.nn as nn
import torchvision.models as models

class CrossAttentionGaze(nn.Module):
    def __init__(self, scene_dim=512, ctx_dim=256, num_heads=8):
        super().__init__()
        self.q_proj=nn.Linear(ctx_dim,scene_dim)
        self.attn=nn.MultiheadAttention(scene_dim,num_heads,batch_first=True,dropout=0.1)
        self.norm=nn.LayerNorm(scene_dim)
    def forward(self,scene_tokens,context):
        q=self.q_proj(context).unsqueeze(1)
        out,_=self.attn(q,scene_tokens,scene_tokens)
        return self.norm(scene_tokens+out.expand_as(scene_tokens))

class HeatmapDecoder(nn.Module):
    """Decoder com skip connection U-Net. Fluxo: 7->14->28(+skip)->56->64"""
    def __init__(self,in_channels=512,skip_channels=512):
        super().__init__()
        self.up1=nn.Sequential(nn.ConvTranspose2d(in_channels,256,4,stride=2,padding=1),
                               nn.BatchNorm2d(256),nn.GELU())
        self.up2=nn.Sequential(nn.ConvTranspose2d(256,128,4,stride=2,padding=1),
                               nn.BatchNorm2d(128),nn.GELU())
        self.skip_proj=nn.Conv2d(skip_channels,128,1)
        self.merge=nn.Sequential(nn.Conv2d(256,128,3,padding=1),nn.BatchNorm2d(128),nn.GELU())
        self.up3=nn.Sequential(nn.ConvTranspose2d(128,64,4,stride=2,padding=1),
                               nn.BatchNorm2d(64),nn.GELU())
        self.up4=nn.Sequential(nn.Conv2d(64,32,3,padding=1),nn.GELU(),
                               nn.Upsample(size=(64,64),mode='bilinear',align_corners=False),
                               nn.Conv2d(32,1,1))
    def forward(self,x,skip):
        x=self.up1(x); x=self.up2(x)
        skip=self.skip_proj(skip)
        x=self.merge(torch.cat([x,skip],dim=1))
        x=self.up3(x); x=self.up4(x)
        B=x.shape[0]
        return F.softmax(x.view(B,-1),dim=1).view(B,1,64,64)

class GazeNetV40(nn.Module):
    def __init__(self,pretrained=True):
        super().__init__()
        bb50=models.resnet50(weights="IMAGENET1K_V2" if pretrained else None)
        self.scene_layer0=nn.Sequential(bb50.conv1,bb50.bn1,bb50.relu,bb50.maxpool)
        self.scene_layer1=bb50.layer1
        self.scene_layer2=bb50.layer2   # skip: 28x28, 512ch
        self.scene_layer3=bb50.layer3
        self.scene_layer4=bb50.layer4
        self.scene_proj=nn.Conv2d(2048,512,1)

        bb18=models.resnet18(weights="IMAGENET1K_V1" if pretrained else None)
        self.head_encoder=nn.Sequential(*list(bb18.children())[:-2])
        self.head_pool=nn.AdaptiveAvgPool2d(1)
        self.head_proj=nn.Linear(512,256)

        self.eye_encoder=nn.Sequential(
            nn.Linear(4,64),nn.GELU(),nn.Linear(64,128),nn.GELU(),nn.Linear(128,256))
        self.ctx_proj=nn.Sequential(nn.Linear(512,256),nn.GELU())
        self.cross_attn=CrossAttentionGaze(512,256,8)
        self.heatmap_decoder=HeatmapDecoder(512,512)
        self._build_coord_grid()

    def _build_coord_grid(self,size=64):
        c=torch.linspace(0,1,size)
        gx,gy=torch.meshgrid(c,c,indexing='xy')
        self.register_buffer('grid_x',gx.unsqueeze(0))
        self.register_buffer('grid_y',gy.unsqueeze(0))

    def soft_argmax(self,heatmap):
        hmap=heatmap.squeeze(1)
        return torch.stack([(hmap*self.grid_x).sum(dim=[1,2]),
                             (hmap*self.grid_y).sum(dim=[1,2])],dim=1)

    def forward(self,scene,head,eye_input):
        B=scene.shape[0]
        s=self.scene_layer0(scene); s=self.scene_layer1(s)
        skip=self.scene_layer2(s)
        s=self.scene_layer3(skip); s=self.scene_layer4(s)
        s=self.scene_proj(s)
        scene_tokens=s.flatten(2).permute(0,2,1)
        h=self.head_pool(self.head_encoder(head)).flatten(1)
        h=self.head_proj(h); e=self.eye_encoder(eye_input)
        ctx=self.ctx_proj(torch.cat([h,e],dim=1))
        attended=self.cross_attn(scene_tokens,ctx)
        feat_map=attended.permute(0,2,1).view(B,512,7,7)
        heatmap=self.heatmap_decoder(feat_map,skip)
        return heatmap, self.soft_argmax(heatmap)

device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
model=GazeNetV40(pretrained=True).to(device)

with torch.no_grad():
    hm,gp=model(torch.randn(2,3,224,224).to(device),
                 torch.randn(2,3,112,112).to(device),
                 torch.randn(2,4).to(device))
    print(f"Heatmap: {hm.shape}  Gaze point: {gp.shape}")
    print(f"Heatmap sum: {hm.squeeze(1).sum(dim=[1,2])}")
print(f"Parametros: {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")
```

### 4.2 Treinamento

```python
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR, LinearLR, SequentialLR
from tqdm.notebook import tqdm

LR=1e-4; EPOCHS=50; EARLY_STOP_PATIENCE=12; WARMUP_EPOCHS=5
GRAD_CLIP=1.0; KL_WEIGHT=2.0; L1_WEIGHT=2.0

backbone_params=(list(model.scene_layer0.parameters())+list(model.scene_layer1.parameters())+
                 list(model.scene_layer2.parameters())+list(model.scene_layer3.parameters())+
                 list(model.scene_layer4.parameters())+list(model.head_encoder.parameters()))
head_params=(list(model.scene_proj.parameters())+list(model.head_proj.parameters())+
             list(model.eye_encoder.parameters())+list(model.ctx_proj.parameters())+
             list(model.cross_attn.parameters())+list(model.heatmap_decoder.parameters()))

optimizer=AdamW([{'params':backbone_params,'lr':LR*0.1},{'params':head_params,'lr':LR}],
                weight_decay=5e-5)
warmup=LinearLR(optimizer,start_factor=0.1,end_factor=1.0,total_iters=WARMUP_EPOCHS)
cosine=CosineAnnealingLR(optimizer,T_max=EPOCHS-WARMUP_EPOCHS,eta_min=1e-6)
scheduler=SequentialLR(optimizer,schedulers=[warmup,cosine],milestones=[WARMUP_EPOCHS])
scaler=torch.amp.GradScaler('cuda')

def gaze_loss(ph,pp,th,tp):
    kl=F.kl_div(ph.squeeze(1).clamp(min=1e-8).log(),th.clamp(min=1e-8),reduction='batchmean')
    return KL_WEIGHT*kl + L1_WEIGHT*F.l1_loss(pp,tp)

def angular_error_deg(pred,target):
    c=torch.tensor([0.5,0.5],device=pred.device)
    cos_a=F.cosine_similarity(pred-c,target-c,dim=1).clamp(-1,1)
    return torch.acos(cos_a).mean().item()*(180/3.14159)

def euclidean_dist(pred,target):
    return ((pred-target)**2).sum(dim=1).sqrt().mean().item()

history={"train_loss":[],"val_loss":[],"val_ang_err":[],"val_euc_dist":[]}
best_val_loss=float('inf'); patience_counter=0

print(f"GazeNet v4.0 | KL={KL_WEIGHT} L1={L1_WEIGHT} | sigma={GazeFollowDataset.SIGMA} | HEAD={HEAD_SIZE}")

for epoch in range(1,EPOCHS+1):
    model.train(); train_loss=0.0; optimizer.zero_grad()
    for step,(scene,head,eye,heatmap,gaze_pt) in enumerate(
            tqdm(train_loader,desc=f"Epoch {epoch}/{EPOCHS} [train]")):
        scene,head,eye=scene.to(device),head.to(device),eye.to(device)
        heatmap,gaze_pt=heatmap.to(device),gaze_pt.to(device)
        with torch.amp.autocast('cuda'):
            ph,pp=model(scene,head,eye)
            loss=gaze_loss(ph,pp,heatmap,gaze_pt)/ACCUM_STEPS
        scaler.scale(loss).backward()
        if (step+1)%ACCUM_STEPS==0 or (step+1)==len(train_loader):
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(),GRAD_CLIP)
            scaler.step(optimizer); scaler.update(); optimizer.zero_grad()
        train_loss+=loss.item()*ACCUM_STEPS
    train_loss/=len(train_loader)

    model.eval(); val_loss=ang_err=euc_dist=0.0
    with torch.no_grad():
        for scene,head,eye,heatmap,gaze_pt in tqdm(test_loader,desc=f"Epoch {epoch}/{EPOCHS} [val]"):
            scene,head,eye=scene.to(device),head.to(device),eye.to(device)
            heatmap,gaze_pt=heatmap.to(device),gaze_pt.to(device)
            with torch.amp.autocast('cuda'):
                ph,pp=model(scene,head,eye); val_loss+=gaze_loss(ph,pp,heatmap,gaze_pt).item()
            ang_err+=angular_error_deg(pp,gaze_pt); euc_dist+=euclidean_dist(pp,gaze_pt)
    val_loss/=len(test_loader); ang_err/=len(test_loader); euc_dist/=len(test_loader)
    scheduler.step()

    history["train_loss"].append(train_loss); history["val_loss"].append(val_loss)
    history["val_ang_err"].append(ang_err); history["val_euc_dist"].append(euc_dist)

    tag=""
    if val_loss<best_val_loss:
        best_val_loss=val_loss; torch.save(model.state_dict(),CHECKPOINT_V4)
        tag=" salvo"; patience_counter=0
    else: patience_counter+=1

    print(f"Epoch {epoch:02d}/{EPOCHS} | Loss:{train_loss:.4f} | Val:{val_loss:.4f} | "
          f"Ang:{ang_err:.2f}graus | Euc:{euc_dist:.4f}{tag}")
    if patience_counter>=EARLY_STOP_PATIENCE:
        print(f"Early stopping epoch {epoch}."); break

print(f"Melhor val loss: {best_val_loss:.4f}")
```

## Modulo 5 -- Deteccao de Objetos
*YOLOv8m, 80 classes COCO*
### 5.1 Funcoes

```python
def detect_all_objects(image_path, conf_threshold=0.4):
    results=yolo_model(image_path,conf=conf_threshold,verbose=False)[0]
    img=Image.open(image_path).convert("RGB"); W,H=img.size
    objects=[]
    for i,box in enumerate(results.boxes):
        cls_id=int(box.cls[0].item()); cls_name=yolo_model.names[cls_id]
        conf=float(box.conf[0].item()); x1,y1,x2,y2=box.xyxy[0].cpu().numpy()
        objects.append({"object_id":i,"class":cls_name,"class_id":cls_id,"confidence":conf,
                        "bbox_pixel":[x1,y1,x2,y2],"bbox_norm":[x1/W,y1/H,(x2-x1)/W,(y2-y1)/H],
                        "center_x":((x1+x2)/2)/W,"center_y":((y1+y2)/2)/H})
    return objects

def find_gazed_objects(gaze_x, gaze_y, objects, radius=0.08):
    gazed=[]
    for obj in objects:
        if obj["class"]=="person": continue
        x1,y1,w,h=obj["bbox_norm"]; cx,cy=obj["center_x"],obj["center_y"]
        dist=((cx-gaze_x)**2+(cy-gaze_y)**2)**0.5
        inside=(x1<=gaze_x<=x1+w) and (y1<=gaze_y<=y1+h)
        if inside or dist<radius: gazed.append({**obj,"dist_to_gaze":dist})
    return sorted(gazed,key=lambda o:o["dist_to_gaze"])

print("Funcoes de deteccao de objetos definidas!")
```

## Modulo 6 -- Fusao: Gaze + Objetos + Pessoas
### 6.1 Carregar Modelo e Inferencia

```python
import torchvision.transforms as T

model=GazeNetV40(pretrained=False).to(device)
model.load_state_dict(torch.load(CHECKPOINT_V4,map_location=device))
model.eval()
print(f"Modelo v4.0 carregado: {CHECKPOINT_V4}")

infer_scene_tf=T.Compose([T.Resize((224,224)),T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])])
infer_head_tf=T.Compose([T.Resize((HEAD_SIZE,HEAD_SIZE)),T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])])

def predict_gaze(img_pil, person):
    W,H=img_pil.size
    hx1,hy1,hx2,hy2=person["head_bbox_pixel"]
    hx1,hy1=max(0,hx1),max(0,hy1); hx2,hy2=min(W,hx2),min(H,hy2)
    head_crop=img_pil.crop((hx1,hy1,hx2,hy2)) if hx2>hx1 and hy2>hy1 else img_pil
    scene_t=infer_scene_tf(img_pil).unsqueeze(0).to(device)
    head_t=infer_head_tf(head_crop).unsqueeze(0).to(device)
    ex,ey=person["eye_x"],person["eye_y"]; off=0.03
    eye_t=torch.tensor([[max(0.,ex-off),ey,min(1.,ex+off),ey]],
                        dtype=torch.float32).to(device)
    with torch.no_grad():
        _,gaze=model(scene_t,head_t,eye_t)
    return float(gaze[0,0].cpu()),float(gaze[0,1].cpu())

def get_image_quadrant(gaze_x,gaze_y):
    col="left" if gaze_x<1/3 else ("center" if gaze_x<2/3 else "right")
    row="top"  if gaze_y<1/3 else ("middle" if gaze_y<2/3 else "bottom")
    return f"{row} {col}"

print("Pipeline de inferencia v4.0 definido!")
```

### 6.2 Analise de Imagem

```python
def analyze_image(image_path, gaze_radius=0.08, verbose=True):
    img,people=detect_people_and_faces(image_path)
    objects=detect_all_objects(image_path)
    results=[]

    if verbose:
        people_str=", ".join(f"Pessoa {p['person_id']}" for p in people) or "nenhuma"
        obj_counts={}
        for o in objects:
            if o["class"]!="person": obj_counts[o["class"]]=obj_counts.get(o["class"],0)+1
        obj_str=", ".join(f"{c} ({n}x)" if n>1 else c for c,n in obj_counts.items()) or "nenhum"
        print(f"  Na imagem ha: {people_str} | objetos: {obj_str}")

    for person in people:
        gaze_x,gaze_y=predict_gaze(img,person)
        gazed_objects=find_gazed_objects(gaze_x,gaze_y,objects,radius=gaze_radius)
        looked_persons=[]
        for other in people:
            if other["person_id"]==person["person_id"]: continue
            x1,y1,w,h=other["bbox_norm"]
            if x1<=gaze_x<=x1+w and y1<=gaze_y<=y1+h:
                looked_persons.append(other["person_id"])
        quadrant=get_image_quadrant(gaze_x,gaze_y)
        result={"person_id":person["person_id"],"gaze_x":gaze_x,"gaze_y":gaze_y,
                "quadrant":quadrant,"gazed_objects":gazed_objects,
                "looked_persons":looked_persons,"eye_source":person.get("eye_source","fallback")}
        results.append(result)

        if verbose:
            parts=[]
            if looked_persons: parts.append(", ".join(f"Pessoa {p}" for p in looked_persons))
            if gazed_objects:  parts.append(", ".join(o["class"] for o in gazed_objects))
            src=">" if person.get("eye_source")=="mediapipe" else "!"
            if parts:
                print(f"  [{src}] Pessoa {person['person_id']} olha para: "
                      f"{' e '.join(parts)} -> ({gaze_x:.3f},{gaze_y:.3f})")
            else:
                print(f"  [{src}] Pessoa {person['person_id']} olha para "
                      f"({gaze_x:.3f},{gaze_y:.3f}) -- nenhum objeto proximo")
    return img,people,objects,results

print("analyze_image v4.0 definida!  [>]=MediaPipe  [!]=fallback")
```

## Modulo 7 -- Descricao de Cena com LLM
### 7.1 Funcao de Descricao

```python
import anthropic, base64

def describe_scene(image_path, results):
    lines=[]
    for r in results:
        if r.get("looked_persons"):
            lines.append(f"Person {r['person_id']} is looking at "
                         f"Person {', '.join(map(str,r['looked_persons']))}.")
        elif r["gazed_objects"]:
            objs=", ".join(o["class"] for o in r["gazed_objects"])
            lines.append(f"Person {r['person_id']} is looking at: {objs} (region: {r['quadrant']}).")
        else:
            lines.append(f"Person {r['person_id']} is looking toward the {r['quadrant']}.")

    with open(image_path,"rb") as f:
        img_b64=base64.standard_b64encode(f.read()).decode("utf-8")
    ext=image_path.rsplit(".",1)[-1].lower()
    media_type="image/jpeg" if ext in ("jpg","jpeg") else f"image/{ext}"

    client=anthropic.Anthropic()
    resp=client.messages.create(
        model="claude-sonnet-4-20250514",max_tokens=300,
        messages=[{"role":"user","content":[
            {"type":"image","source":{"type":"base64","media_type":media_type,"data":img_b64}},
            {"type":"text","text":(
                f"Gaze analysis results:\n{''.join(lines)}\n\n"
                "Based on the image and the gaze information above, "
                "write a single short paragraph (2-3 sentences) describing "
                "what is happening in this scene. Be concise and factual."
            )}
        ]}]
    )
    return resp.content[0].text

print("describe_scene definida!")
```

```python
from tqdm.notebook import tqdm as tqdm_nb

N_EVAL=500
df_eval=df_test.drop_duplicates(subset=["image_path","eye_x","eye_y"]).head(N_EVAL).reset_index(drop=True)

results_classic={"ang_err":[],"euc_dist":[],"failed":0}
results_modern ={"ang_err":[],"euc_dist":[],"failed":0}
model.eval()

eval_scene_tf=T.Compose([T.Resize((224,224)),T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])])
eval_head_tf=T.Compose([T.Resize((HEAD_SIZE,HEAD_SIZE)),T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])])

def ang_err_single(px,py,gx,gy):
    cx=cy=0.5
    p=np.array([px-cx,py-cy]); t=np.array([gx-cx,gy-cy])
    pn=np.linalg.norm(p); tn=np.linalg.norm(t)
    if pn<1e-6 or tn<1e-6: return 90.0
    return np.degrees(np.arccos(np.clip(np.dot(p,t)/(pn*tn),-1,1)))

def euc_dist_single(px,py,gx,gy): return ((px-gx)**2+(py-gy)**2)**0.5

print(f"Avaliando {N_EVAL} imagens do GazeFollow test set...")

for _,row in tqdm_nb(df_eval.iterrows(),total=N_EVAL):
    img_path=os.path.join(DATASET_ROOT,row["image_path"].strip())
    gt_x,gt_y=row["gaze_x"],row["gaze_y"]
    try:
        img_pil=Image.open(img_path).convert("RGB"); W,H=img_pil.size
        x1=max(0,int(row["x_min"]*W)); y1=max(0,int(row["y_min"]*H))
        x2=min(W,int((row["x_min"]+row["w"])*W)); y2=min(H,int((row["y_min"]+row["h"])*H))

        faces=get_face_and_eyes_mediapipe(img_pil)
        fc_x=row["x_min"]+row["w"]/2; fc_y=row["y_min"]+row["h"]/2
        face_match=None
        for face in faces:
            fx,fy,fw,fh=face["face_bbox_norm"]
            if fx<=fc_x<=fx+fw and fy<=fc_y<=fy+fh: face_match=face; break

        if face_match:
            eye_x,eye_y=face_match["eye_x"],face_match["eye_y"]; eye_src="mediapipe"
            fx,fy,fw,fh=face_match["face_bbox_norm"]; m=0.1
            hx1=max(0,int((fx-fw*m)*W)); hy1=max(0,int((fy-fh*m)*H))
            hx2=min(W,int((fx+fw*(1+m))*W)); hy2=min(H,int((fy+fh*(1+m))*H))
        else:
            eye_x=row["x_min"]+row["w"]/2; eye_y=row["y_min"]+row["h"]*0.20
            eye_src="fallback"; hx1,hy1,hx2,hy2=x1,y1,x2,y2

        person_dict={"person_id":1,"eye_x":eye_x,"eye_y":eye_y,"eye_source":eye_src,
                     "head_bbox_pixel":[hx1,hy1,hx2,hy2],"bbox_pixel":[x1,y1,x2,y2],
                     "bbox_norm":[row["x_min"],row["y_min"],row["w"],row["h"]]}

        # Classico
        cr=estimate_gaze_classic(img_pil,person_dict)
        if cr:
            results_classic["ang_err"].append(ang_err_single(cr[0],cr[1],gt_x,gt_y))
            results_classic["euc_dist"].append(euc_dist_single(cr[0],cr[1],gt_x,gt_y))
        else: results_classic["failed"]+=1

        # Moderno
        head_crop=img_pil.crop((hx1,hy1,hx2,hy2)) if hx2>hx1 and hy2>hy1 else img_pil
        st=eval_scene_tf(img_pil).unsqueeze(0).to(device)
        ht=eval_head_tf(head_crop).unsqueeze(0).to(device)
        off=0.03
        et=torch.tensor([[max(0.,eye_x-off),eye_y,min(1.,eye_x+off),eye_y]],
                         dtype=torch.float32).to(device)
        with torch.no_grad():
            _,gaze=model(st,ht,et)
        mx,my=float(gaze[0,0].cpu()),float(gaze[0,1].cpu())
        results_modern["ang_err"].append(ang_err_single(mx,my,gt_x,gt_y))
        results_modern["euc_dist"].append(euc_dist_single(mx,my,gt_x,gt_y))

    except Exception as e:
        results_classic["failed"]+=1; results_modern["failed"]+=1

print("Avaliacao concluida!")
```
